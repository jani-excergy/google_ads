```python

my_df['f2']=np.sqrt(my_df['Completed_Jobs']**3)*np.abs(my_df['Avg_Sales'])
my_df['f3']=np.abs(np.sqrt(my_df['Completed_Jobs']) - np.sqrt(my_df['tech_count']))
my_df['f4']=np.log(np.sqrt(my_df['Completed_Jobs'])*my_df['TMAX']**3)
```

```python
my_df=my_df.drop_duplicates(ignore_index=True)
```

```python
train=my_df[(my_df['ds']>='2019-01-03') & (my_df['ds']<='2020-12-31')]
```

```python
test=my_df[(my_df['ds']>'2020-12-31')]
```

```python
train=train.drop_duplicates(ignore_index=True)
```

```python
m=Prophet(interval_width=0.95)

```

```python
train.columns
```

```python
m.add_regressor('TMAX',standardize=False)
m.add_regressor('Completed_Jobs',standardize=False,mode='multiplicative')
m.add_regressor('tech_count',standardize=False,mode='multiplicative')




```

```python
m.fit(train)
```

```python
m.params
```
