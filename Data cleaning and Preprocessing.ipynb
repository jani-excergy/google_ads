```python
# sorting the busineed data by created date
df1= df1.sort_values('Created Date')
```

```python
# Ranking the groups based on income

df1["group_rank"] = df1.groupby("Created Date")["Income"].rank(ascending=1,method='dense')


```

```python
df1.head(100)
```

```python
# grouping created date based on max group rank
x=df1.groupby('Created Date').group_rank.max()
```

```python

x=pd.DataFrame(x)
```

```python
x.dtypes
```

```python
# converting object into dataframe

a=x.add_suffix('').reset_index()
```

```python
a.head()
```

```python
# merging business data and Ads by imperssion data 
mer= pd.merge(df1,a,on=['Created Date','group_rank'])
```
