```python
import pandas as pd
```

```python
train= df[0:710]
test= df[710:]
```

```python
X=train[0:650]
```

```python
X.head()
```

```python
X=X.drop(['Unnamed: 0'],axis=1)
```

```python
X.head()
```

```python
y_train= X['Income']
X_train= X.drop(['Income'],axis=1)
```

```python
X_train['Day']=pd.to_datetime(X_train['Day'])
```

```python
X_train["day"] = pd.to_datetime(X_train.Day, format="%y-%m-%d").dt.day
X_train["month"] = pd.to_datetime(X_train.Day, format="%y-%m-%d").dt.month
X_train["year"] = pd.to_datetime(X_train.Day, format="%y-%m-%d").dt.year 
```

```python
X_train= X_train.drop(['Day'],axis=1)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import train_test_split
```

```python
reg= ExtraTreesRegressor()
reg.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
```

```python
feat_importances = pd.Series(reg.feature_importances_, index=X_train.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from scipy.stats import randint


parameters = {'n_estimators': randint(10,1000),
               'max_features': ['auto','sqrt'],
               'max_depth': randint(3,40),
               'min_samples_split': randint(2,30),
               'min_samples_leaf': randint(1,10)  }
```

```python
base_learner=RandomForestRegressor()
```

```python
tuned_model= RandomizedSearchCV(estimator = base_learner, param_distributions = parameters,scoring='neg_mean_squared_error', n_iter = 100, cv = 5, verbose=2, random_state=42)
```

```python
tuned_model.fit(new_df,y_train)
```

```python
tuned_model.best_params_
```

```python
tuned_model.best_score_
```

```python
tuned_learner=RandomForestRegressor(max_depth=4,max_features= 'auto',min_samples_leaf= 5,min_samples_split= 6,n_estimators=169)
```

```python
tuned_learner.score(new_df,y_train)
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(tuned_learner, new_df, y_train, cv=5)
```
